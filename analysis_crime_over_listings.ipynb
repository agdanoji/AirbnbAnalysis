```python
# average no of bookings done for each months 
#  for low crime areas
X =np.array([15159.740000000027,15196.730000000036,14823.690000000026,13705.930000000015,13964.460000000036,13690.510000000011,13222.89000000004,12615.420000000002,12065.45999999999,11926.369999999893,12631.87999999991,13552.429999999884])
#  for high crime areas 
Y =np.array([8094.569999999979,7912.110000000001,7577.640000000024,7099.559999999997,7273.570000000038,7137.159999999981,7022.4900000000125,7054.259999999984,6674.430000000027,6677.230000000005,7236.4199999999955,7459.149999999995])
```

```python
# Apply permutation test for data obtained over 12 months 
len(X)
```

```python
len(Y)
```

```python
#calculate variances
x_var = np.var(X)
y_var = np.var(Y)
print(x_var)
print(y_var)
```

```python
#calculate means
x_mean = np.mean(X)
y_mean = np.mean(Y)
print(x_mean)
print(y_mean)
print(x_mean-y_mean)
```

```python
se_hat = np.sqrt(x_var/len(X) + y_var/len(Y))
print(x_var/len(X))
print(y_var/len(Y))
print(se_hat)
```

```python
w = np.divide((x_mean - y_mean),se_hat)
print(w)
```

```python
if abs(w)>1.96:
    print("w = "+str(w))
    print("Null Hypothesis Rejected")
else:
    print("w = "+str(w))
    print("Null Hypothesis Accepted")
```

```python
#alpha = 0.05
pValue = np.multiply(2,0)
print(pValue)
```

```python
#Permutation Test: Part(b)
t_obs = abs(x_mean-y_mean)
Z = np.concatenate((X,Y))
count = 0
perm = np.random.permutation(Z)
x_new = perm[0:10]
y_new = perm[10:20]
for i in range(0,1000000):
    perm = np.random.permutation(Z)
    x_new = perm[0:10]
    y_new = perm[10:20]
    x_new_mean = np.mean(x_new)
    y_new_mean = np.mean(y_new)
    t_i = abs(x_new_mean - y_new_mean)
    if t_i>t_obs:
        count+=1

pval = count/1000000
print(count)
print("pval is "+str(pval))
    
```

```python
if pval<0.05:
    print("Null Hypothesis Rejected")
else:
    print("Null Hypothesis Accepted")
```
